```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir los valores de y en el rango relevante
y_values = np.linspace(-1, 3, 500)
```

```python
# Definir f(y) y F(y) usando las condiciones dadas
f_y = np.where((y_values >= 0) & (y_values <= 2), (1/2) * y_values, 0)
F_y = np.piecewise(y_values, [y_values < 0, (y_values >= 0) & (y_values <= 2), y_values > 2],
                   [0, lambda y: y**2 / 4, 1])
```

```python
# Graficar f(y) y F(y)
plt.figure(figsize=(12, 5))
# Subplot para f(y)
plt.subplot(1, 2, 1)
plt.plot(y_values, f_y, label='$f(y)$', color='blue')
plt.title('$f(y)$ - Función de Densidad')
plt.xlabel('$y$')
plt.ylabel('$f(y)$')
plt.legend()

# Subplot para F(y)
plt.subplot(1, 2, 2)
plt.plot(y_values, F_y, label='$F(y)$', color='green')
plt.title('$F(y)$ - Función de Distribución Acumulada')
plt.xlabel('$y$')
plt.ylabel('$F(y)$')
plt.legend()

plt.tight_layout()
plt.show()
```
