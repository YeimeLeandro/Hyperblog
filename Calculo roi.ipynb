```python
# Código Python para Proyección de Ventas Mensuales y Cálculo de ROI

# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Configurar el estilo de los gráficos
plt.style.use('seaborn')

# Función para formatear números como moneda
def formato_moneda(valor):
    return f"${valor:,.2f}"

# Función para formatear porcentajes
def formato_porcentaje(valor):
    return f"{valor:.2f}%"

# Título del Programa
print("=============================================================")
print("  Proyección de Ventas Mensuales y Cálculo de ROI con IA  ")
print("=============================================================\n")

# -------------------- Entradas del Usuario --------------------
print("=== Entradas del Usuario ===\n")

try:
    # Inversión Inicial
    inversion_inicial = float(input("Ingrese la inversión inicial en IA ($): "))

    # Costos Operativos Mensuales
    costos_operativos_mensuales = float(input("Ingrese los costos operativos mensuales de IA ($): "))

    # Ventas Iniciales Mensuales
    ventas_iniciales = float(input("Ingrese las ventas iniciales mensuales antes de implementar IA ($): "))

    # Tasa de Crecimiento Mensual de Ventas (%)
    tasa_crecimiento_mensual = float(input("Ingrese la tasa de crecimiento mensual de ventas debido a la IA (%): "))

    # Periodo de Proyección
    periodo_proyeccion = int(input("Ingrese el periodo de proyección (en meses): "))

except ValueError:
    print("Por favor, ingrese valores numéricos válidos.")
    exit()

# Convertir tasa de crecimiento a decimal
tasa_crecimiento_mensual_decimal = tasa_crecimiento_mensual / 100

# -------------------- Cálculos de Proyección --------------------
print("\n=== Realizando Cálculos de Proyección ===\n")

# Crear una lista para almacenar los datos mensuales
meses = []
ventas_proyectadas = []
beneficios_netos = []

# Inicializar ventas del primer mes
ventas_mes_anterior = ventas_iniciales

for mes in range(1, periodo_proyeccion + 1):
    meses.append(mes)

    # Calcular ventas del mes actual
    ventas_mes_actual = ventas_mes_anterior * (1 + tasa_crecimiento_mensual_decimal)
    ventas_proyectadas.append(ventas_mes_actual)

    # Calcular beneficios netos
    beneficio_neto = ventas_mes_actual - costos_operativos_mensuales
    beneficios_netos.append(beneficio_neto)

    # Actualizar ventas del mes anterior para el siguiente ciclo
    ventas_mes_anterior = ventas_mes_actual

# Crear un DataFrame con los resultados
df = pd.DataFrame({
    'Mes': meses,
    'Ventas Proyectadas ($)': ventas_proyectadas,
    'Beneficios Netos ($)': beneficios_netos
})

# Calcular beneficios totales
beneficios_totales = df['Beneficios Netos ($)'].sum()

# Calcular ROI
roi = (beneficios_totales / inversion_inicial) * 100

# -------------------- Mostrar Resultados --------------------
print("=== Resultados de la Proyección ===\n")

# Mostrar el DataFrame
display(df)

# Resumen de Beneficios y ROI
print(f"**Beneficios Totales a lo Largo del Periodo de Proyección:** {formato_moneda(beneficios_totales)}")
print(f"**Inversión Inicial en IA:** {formato_moneda(inversion_inicial)}")
print(f"**ROI:** {formato_porcentaje(roi)}\n")

# -------------------- Visualización de Datos --------------------
print("=== Visualización de la Proyección ===\n")

# Gráfico de Ventas Proyectadas
plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Ventas Proyectadas ($)'], marker='o', label='Ventas Proyectadas')
plt.title('Proyección de Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas ($)')
plt.grid(True)
plt.legend()
plt.show()

# Gráfico de Beneficios Netos
plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Beneficios Netos ($)'], marker='s', color='green', label='Beneficios Netos')
plt.title('Beneficios Netos Mensuales')
plt.xlabel('Mes')
plt.ylabel('Beneficios Netos ($)')
plt.grid(True)
plt.legend()
plt.show()

# -------------------- Conclusión --------------------
print("=============================================================")
print("¡Gracias por usar el Proyector de Ventas y Calculadora de ROI!")
print("=============================================================")
```
